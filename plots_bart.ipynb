```python
import os
from os import listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
import regex as re
```

```python
sim_folders = [x[0] for x in os.walk('dummy_bart\enemy 1')][1:]
```

```python
datafiles = [[sim_folder + '\\' + file for file in listdir(sim_folder) if isfile(join(sim_folder, file))] for sim_folder in sim_folders]
datafiles
```

```python
means_sims = [np.zeros(21), np.zeros(21), np.zeros(21), np.zeros(21)]
```

```python
# best per generation
index = [0,2,3,5]
for sim_folder in datafiles:
        for i in range(len(index)):
            datafile = sim_folder[index[i]]
            data = np.loadtxt(datafile)[-21:]
#             print(data)
            means_sims[i] += data/11
```

```python
for mean in means_sims:
    plt.plot(range(len(mean)), mean)
    plt.show()
```
