```python
import os
from os import listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
import regex as re
```

```python
sim_folders = [x[0] for x in os.walk('dummy_bart\enemy 2')][1:]
```

```python
datafiles = [[sim_folder + '\\' + file for file in listdir(sim_folder) if isfile(join(sim_folder, file))] for sim_folder in sim_folders]
datafiles
```

```python
means_sims = [np.zeros(21), np.zeros(21), np.zeros(21), np.zeros(21)]
titles = ['']
last_mean = []
```

```python
font = {'family' : 'normal',
        'size'   : 18}

plt.rc('font', **font)
```

```python
# best per generation
index = [2]
for sim_folder in datafiles:
        for i in range(len(index)):
            datafile = sim_folder[index[i]]
            data = np.loadtxt(datafile)[-21:]
            data=data[-1]
            last_mean.append(data)
```

```python
last_mean
```

```python
fig, ax = plt.subplots(3)
fig.suptitle('Average per generation over 10 simulations vs enemy 3')
fig.set_size_inches(18,10)
ax[0].set_ylabel('Fitness')
ax[1].set_ylabel('Fitness')
ax[2].set_ylabel('Lifepoints')

ax[0].plot(range(len(means_sims[1])),means_sims[1])
ax[0].set_ylabel('Average Fitness')

ax[1].plot(means_sims[-1])
ax[1].set_ylabel('Std')
ax[2].plot(means_sims[2])
ax[2].set_ylabel('Lifepoints')

ax[2].set_xlabel('Generations')

fig.savefig('average_simulations_en3.png')

# ax[0].set_title('Fitness')
# ax[1].set_title('Fitness')
# ax[2].set_title('STD')

# for mean in means_sims[1:]:
#     ax[mean].plot(range(len(means_sims[mean])), means_sims[mean])
```
